```python
eeg_df.columns=['e1','e2','e3','t7','c3','c4','t8']
```

```python
eeg1=eeg_df.loc[:,["t7"]]
eeg2=eeg_df.loc[:,["c3"]]
eeg3=eeg_df.loc[:,["c4"]]
eeg4=eeg_df.loc[:,["t8"]]
```

```python
s_f=len(eeg1)/120
t_f=1/s_f
T=np.arange(0,120,t_f)
```

```python
len(T)
```

```python
plt.Figure(figsize= (10,8))
plt.plot(T,eeg1,'r')
plt.ylabel('V')
plt.xlabel('Time-s')
plt.title('EEG T7')
plt.show()
```

```python
plt.Figure(figsize= (10,8))
plt.plot(T,eeg2,'b')
plt.ylabel('V')
plt.xlabel('Time-s')
plt.title('EEG C3')
plt.show()
```

```python
plt.Figure(figsize= (10,8))
plt.plot(T,eeg3,'g')
plt.ylabel('V')
plt.xlabel('Time-s')
plt.title('EEG C4')
plt.show()
```

```python
plt.Figure(figsize= (10,8))
plt.plot(T,eeg4,'k')
plt.ylabel('V')
plt.xlabel('Time-s')
plt.title('EEG T8')
plt.show()
```

```python
E1=np.fft.fft(eeg1)
E2=np.fft.fft(eeg2)
E3=np.fft.fft(eeg3)
E4=np.fft.fft(eeg4)

```

```python
N=np.arange(len(E1))
T_f=len(E1)/s_f
freq=N/T_f
```

```python

plt.Figure(facecolor='#94F008')
plt.subplot(211)

plt.plot(T,eeg1,'r')
plt.ylabel('V')
plt.xlabel('Time-s')
plt.title('EEG T7')


plt.subplot(212)

plt.plot(freq, np.abs(E1), '')
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |T7(freq)|')
plt.xlim(0, 20)
plt.tight_layout()
plt.show()

```
